```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

### 1. Загрузка данных

```python
df = pd.read_csv('data/bank/bank.csv', sep=';')

print(f"Размер исходного датасета: {df.shape}")
df.head(3)
```

### 2. Обработка данных

```python
# Дискретизация признака age (Биннинг)
# Разбиваем возраст на 3 категории: до 25 (молодежь), 25-60 (взрослые), старше 60 (пенсионеры)
bins = [0, 25, 60, 120]
labels = ['young', 'adult', 'senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# Удаляем исходный числовой признак age, так как заменили его на категории
df = df.drop(columns=['age'])

# Преобразование целевой переменной y
# Переводим строковые 'yes'/'no' в машинный бинарный формат 1/0
df['y'] = df['y'].map({'yes': 1, 'no': 0})
```

```python
num_cols = ['balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
cat_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'age_group']

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
# Кодирование категориальных признаков (One-Hot Encoding)
# Параметр drop_first=True удаляет первый столбец каждой категории, 
# чтобы алгоритмы не столкнулись с проблемой мультиколлинеарности (dummy variable trap)
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)

# Проверяем итоговый вид подготовленных данных
print(f"Итоговое количество столбцов после кодирования: {df.shape[1]}")
display(df.head(3))
```
