```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

### 1. Загрузка данных

```python
df = pd.read_csv('data/bank/bank-full.csv', sep=';')

print(f"Размер исходного датасета: {df.shape}")
df.head(3)
```

### 2. Обработка данных

```python
bins = [0, 25, 60, 120]
labels = ['young', 'adult', 'senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
df = df.drop(columns=['age'])
df['y'] = df['y'].map({'yes': 1, 'no': 0})
```

```python
num_cols = ['balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
cat_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'age_group']

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df = pd.get_dummies(df, columns=['job', 'marital', 'education', 'default', 'housing', 'loan',
                                  'contact', 'month', 'poutcome', 'age_group'], drop_first=True)

print(f"Итоговое количество столбцов после кодирования: {df.shape[1]}")
display(df.head(3))
```
